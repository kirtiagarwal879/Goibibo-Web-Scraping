```python
import requests
import pandas as pd  
from bs4 import BeautifulSoup
from tqdm import tqdm

Hotel_name = []
city = []
country = []
Star_rating = []
price_per_night = []
Amenities = []
Hotel_description = []
Customer_review1 = []
Customer_review2 = []
list_of_city = ['kota','shimla','prayagraj','agra','kanpur','jhansi']
```

```python
for i in list_of_city:
    
    URL = 'https://www.goibibo.com/hotels/hotels-in-'+ i +'-ct/'
    
    r = requests.request("GET",URL)
    content = BeautifulSoup(r.text,'html.parser')
    
    hotel_information = content.find_all('a',attrs={'class','HotelCardstyles__HotelNameSeoAnchor-sc-1s80tyk-16 eUMfTF'})
    hotel_url = []
    for h in hotel_information:
        Hotel_name.append(h.get_text())
        city.append(i)
        country.append('India')
        hotel_url.append(h['href'])
        
        murl = []
        
                
        for x in hotel_url:
            murl.append('https://www.goibibo.com' + x)
            
        for y in tqdm(murl):
            req = requests.request("GET",y)
            about = BeautifulSoup(req.text,'html.parser')
            
            
            rate = about.find_all('span',attrs={'class','ReviewInfostyles__ReviewNumber-sc-1hepu7p-3 vmmxn'})
            price = about.find_all('p',attrs={'class','BookingWidgetstyles__PriceValueStyled-sc-1tsb1-4 gpxhWi'})
            desc = about.find('div',attrs={'class','AboutPropertiesstyles__MainContentTextStyle-sc-4xqpe7-2 knLjJo'})
            amn = about.find_all('div',attrs={'class','Amenitiesstyles__AmenitiesListBlock-sc-10opy4a-5 gyoAXQ'})
            rev = about.find_all('span',attrs={'class','UserReviewstyles__UserReviewTextStyle-sc-1y05l7z-4 cHVFbF'})

            for r in rate:
                 Star_rating.append(r.get_text())
                    
            for p in price:
                price_per_night.append(p.get_text())
                
            for am in amn:
                arr = []
                for lis in am.find_all('span',attrs={'class','Amenitiesstyles__AmenityItemText-sc-10opy4a-9 eLihWE'}):
                    arr.append(lis.get_text())
                Amenities.append(arr)
                
            review1 = 'Review is not available'
            review2 = 'Review is not available'
            if(len(rev) > 1):
                review1 = rev[0].text.rstrip()
            
            if(len(rev) > 3):
                review2 = rev[3].text.rstrip()
            
            #print(review1)
            #print(review2)
            Customer_review1.append(review1)
            Customer_review2.append(review2)
            
            
        
```

```python
hdata = pd.DataFrame(list(zip(Hotel_name,Star_rating,city,country,price_per_night,Amenities,Customer_review1,Customer_review2)),columns=['Hotel_name','Star_rating','City','Country','Price_per_night','Amenities','Customer_review1','Customer_review2'])
```

```python
hdata.to_csv('hoteldata.csv')
```
